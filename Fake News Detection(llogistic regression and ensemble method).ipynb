```python
import pandas as pd
import numpy as np

```

```python
true = pd.read_csv('/content/True.csv', engine='python', on_bad_lines='skip')
fake = pd.read_csv('/content/Fake.csv', engine='python', on_bad_lines='skip')
```

```python
true['label'] = 1
fake['label'] = 0
```

```python
true.head()
```

```python
###now concatenating both the tables

news  = pd.concat([true, fake] , axis = 0)
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis = 1)
```

```python
###suffling the data

news = news.sample(frac = 1)
```

```python
news.reset_index(inplace = True)
```

```python
news.head()
```

```python
news.drop(['index'], axis = 1, inplace = True)
```

```python
import re
```

```python
def wordopt(text):

  text = text.lower()

  text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
  text = re.sub(r'<.*?>', '', text)

    # Remove punctuation
  text = re.sub(r'[^\w\s]', '', text)

    # Remove digits
  text = re.sub(r'\d+', '', text)

    # Remove newline characters
  text = re.sub(r'\n', ' ', text)

  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()

LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_lr))
```

```python
### Using other ML models

from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()

DTC.fit(xv_train, y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dtc))
```
